Strings and Slice

```python
const = 'hello'
const = 'y' + const[1:]  # Replace the first character with 'y'
print(const)
```

```python
const = 'abcdefghijklmnop'
for char in const:
    if char in 'aeiou':
        print(char)
    else:
        print('Not a vowel')
```

```python
const = 'abcdefghijklmnop'
print(const[3:5])
print(const[::-1])
print(const[1:len(const)])
```

Algorithms

```python
cube = 8
for guess in range(abs(cube) + 1):
    if guess**3 >= abs(cube):
        print(f'The cube root of {cube} is {guess}.')
        break
else:    print(f'{cube} is not a perfect cube.')
```
