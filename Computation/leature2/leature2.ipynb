For loops

```python
mySum = 0
for i in range(7, 10):
    print("Adding", i)
    mySum += i
print("The sum is:", mySum)
```

```python
sum = 0
for i in range(9, 12):
    sum+=i
    if sum == 19:
        print("Sum reached 19, breaking loop.")
        break
    else:
        print("Current sum:", sum)
print("Sum is:", sum)
```
